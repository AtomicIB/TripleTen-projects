# Encontrar el número de libros publicados después del 1 de enero de 2000.

```python
query1 = """
SELECT COUNT(book_id)
FROM books
WHERE publication_date > '2000-01-01';
"""

pd.io.sql.read_sql(query1, con = engine)
```

Se han publicado **821** libros posterior al 1 de enero del 2000.

# Encontrar el número de reseñas de usuarios y la calificación promedio para cada libro.

```python
query2 = """
SELECT
    books.book_id,
    books.title,
    COUNT(DISTINCT reviews.review_id) AS num_reviews,
    AVG(ratings.rating) AS avg_rating
FROM books
LEFT JOIN reviews ON books.book_id = reviews.book_id
LEFT JOIN ratings ON books.book_id = ratings.book_id
GROUP BY books.book_id
ORDER BY num_reviews DESC, avg_rating DESC;

"""

pd.io.sql.read_sql(query2, con = engine)
```

Generamos una tabla con la cantidad de reseñas y la calificación promedio de cada libro, podemos ver que el libro con más reseñas fue Twilight con una calificación promedio de 3.6 y que tenemos varios libros sin ninguna reseña pero que aún así tienen calificación promedio. Seguramente este servicio ocupa reseñas externas para su campor de calificación promedio.

# Identificar la editorial que ha publicado el mayor número de libros con más de 50 páginas.

```python
query3 = """
SELECT
    publishers.publisher as publisher,
    COUNT( books.book_id) as num_books
FROM publishers
INNER JOIN books on books.publisher_id = publishers.publisher_id
WHERE books.num_pages > 50
GROUP BY publisher
ORDER BY num_books DESC
LIMIT 1;

"""

pd.io.sql.read_sql(query3, con = engine)
```

La editorial que ha publicado más libros (con más de 50 páginas) es Penguin Books con 42 libros.

# Identificar al autor que tiene la más alta calificación promedio del libro: solo los libros con al menos 50 calificaciones.

```python
query4 = """
SELECT 
    authors.author as author_name,
    AVG(ratings.rating) as avg_rating
FROM authors
INNER JOIN books on books.author_id = authors.author_id
INNER JOIN ratings on ratings.book_id = books.book_id
GROUP BY author_name
HAVING COUNT(ratings.rating) > 50
ORDER BY avg_rating DESC
LIMIT 1;
"""

pd.io.sql.read_sql(query4, con = engine)
```

El autor con la calificacion promedio más alta considerando solo los libros con al menos 50 calificaciones es J.K. Rowling/Mary GrandPré.

# Encontrar el número promedio de reseñas de texto entre los usuarios que calificaron más de 50 libros.

```python
query5 = """
SELECT AVG(reviews_cnt) AS avg_reviews_per_user
FROM 
(
  SELECT username, COUNT(*) AS reviews_cnt
  FROM reviews
  GROUP BY username
) AS reviews_per_user
WHERE username IN 
(
  SELECT username
  FROM ratings
  GROUP BY username
  HAVING COUNT(*) > 50
);
"""

pd.io.sql.read_sql(query5, con = engine)
```

El promedio de reseñas entre los usuarios que han calificado más de 50 libros es de 24.33
